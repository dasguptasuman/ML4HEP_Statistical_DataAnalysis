```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
from scipy.stats import norm
random_numbers = np.random.binomial(10,.5,300)
```

```python
p = np.linspace(0,1,500)
```

```python
prior =[]
for k in p :  
    prio = norm.pdf(k, loc = .2, scale = .1)
    prior.append(prio)
plt.plot(p,prior)
plt.show()

k = 0

for i in random_numbers:
    posterior = []
    h = 0
    total = 0
    avg = 0
    for j in p:
        p_post = comb(10,i)*(j**i)*(1-j)**(10-i)*prior[h]
        posterior.append(p_post)
        avg = avg + p_post*j
        total = total + p_post
        h=h+1
    if abs((avg/total - .5))< .01:
        break
    print(len(prior))
    prior = posterior
   # print(avg/total)
    k = k+1;
print("number of steps after which we get actual distribution  : " ,k-1)
plt.plot(p,posterior)
plt.show()
    
```
